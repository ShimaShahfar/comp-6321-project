```python
import Augmentor
import matplotlib.pylab as plt
from pathlib import PosixPath
%matplotlib inline
```

```python
DATA_PATH = PosixPath("../data/membrane/train")
```

```python
ls {DATA_PATH}
```

```python
p = Augmentor.Pipeline(DATA_PATH/"image", output_directory="../aug")
p.ground_truth(DATA_PATH/"label")


p.rotate(probability=0.8, max_left_rotation=15, max_right_rotation=15)
p.flip_left_right(probability=0.5)
p.flip_top_bottom(probability=0.5)
p.zoom_random(probability=0.5, percentage_area=0.8)
```

```python
g = p.keras_generator(batch_size=128)
images, labels = next(g)
```
